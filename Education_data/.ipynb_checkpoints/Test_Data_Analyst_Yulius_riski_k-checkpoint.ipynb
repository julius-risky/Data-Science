# 1.  Rancangan klasifikasi yang mempresentasikan tema - tema kalimat 

Dalam membuat klasifikasi text yang mempresentasikan tema - tema kalimat kita harus memerlukan beberapa step umum yang harus dilakukan, antara lain:
 <br>`1. Data Collection`: Data collection adalah hal yang pertama kita lakukan untuk mempersiapkan klasifikasi tema-tema kalimat, data collection merupakan proses mengumpulkan data dan menggali informasi dari informasi data text chat.
 <br>`2. Menyiapkan Data Training`: Sebelum kita melakukan klasifikasi tema-tema kalimat kita perlu membuat data training yang berlabel dari data text chat sebelumnya. Misalnya seperti tabel dibawah ini. 
<br>

| Text | Tema | 
| ----------- | :---------: | 
| Hp gua udah 4g gausah diaktifin lagi |Teknologi |
| Saya ingin menyamoaikan keluhan|Keluhan |
| Mana yg dipilih? |Pertanyaan |

<br> `3. Data Preprocessing`: informasi yang akan digali berisi data-data yang strukturnya “sembarang” atau tidak terstruktur. Oleh karena itu, diperlukan proses pengubahan bentuk menjadi data yang terstruktur. Rancangan text preprocesing dapat dilakukan dengan 1). Case folding, 2). Tokenization, 3). filtering dengan menghapus stop words, 4). spell check, 5). Stemming. untuk penjelasan lebih jauh dapat dilihat pada rancangan data preprocessing. 
<br> `4. Vectorizing`: Dalam tahapan vectorizing ini kita akan membuat suatu kata dalam kalimat menjadi sebuah vektor. Dalam kasus ini saya menyarankan menggunakan metode Vectorizing TFIDF N-gram. 
<br> `5. Modeling`: Setelah kita mengubah kata-kata dalam sebuah kalimat menjadi vector maka langkah selanjutnya melakukan proses modeling atau training data menggunakan algoritma machin learning atau deep learning. Dalam kasus akan menggunakan multiclass text classification ada beberapa methode yang bisa kita gunakan antara lain :
<br>

<br> `6. Evaluation`:Untuk mengetahui model yang digunakan sudah atau tidak kita dapat melakukan evaluasi sehingga dapat meningkatkan performa dari model klasifikasi, karena pada kasus ini kita melakukan klasifikasi metode evaluasi yang dapat digunakan adalah **confusion matrix** dengan melihat akurasinya.
<br>Dari semua algoritma tersebut yang memiliki akurasi terbaik adalah LR_WordLevel_TF-IDF dengan akurasi 0.875 menurut [sumber](https://towardsdatascience.com/model-selection-in-text-classification-ac13eedf6146) akan tetapi berbeda data yang digunakan akan berbeda juga akurasi yang dihasilkan maka dari itu kita perlu mencoba menggunakan beberapa model,untuk pertama jika saya mengerjakan kasus tersebut saya akan terlebih dahulu menggunakan model yang ringan, cepat dan memiliki akurasi yang bagus seperti algoritma XGBOOST. jika dirasa algoritma tersebut masih kurang akurat kita dapat melakukan hyperparameter tuning, Featur engginering, atau mengganti algoritma untuk meningkatkan akurasi model.

## a. Case Folding
Case folding adalah salah satu bentuk text preprocessing yang paling sederhana dan efektif meskipun sering diabaikan. Tujuan dari case folding untuk mengubah semua huruf dalam dokumen menjadi huruf kecil. Hanya huruf ‘a’ sampai ‘z’ yang diterima. Karakter selain huruf dihilangkan dan dianggap delimiter

###  Mengubah text menjadi lowercase

```python
text = text.lower()
text
```

### Menghapus tanda baca

```python
punctuation
```

```python
text = text.translate(str.maketrans("","",punctuation))
text
```

### Menghapus white space

```python
text = text.strip()
text
```

## b. Tokenization
Tokenizing adalah proses pemisahan teks menjadi potongan-potongan yang disebut sebagai token untuk kemudian di analisa. Kata, angka, simbol, tanda baca dan entitas penting lainnya dapat dianggap sebagai token. 

### Memisahkan menjadi kalimat

```python
kalimat = text.split(sep='\n')
kalimat
```

### Memisahkan berdasarkan kata

```python
tokens = word_tokenize(text)
tokens
```

```python
tokens = [token for token in tokens if token.isalnum()]
tokens
```

## c. Filtering (Stopword Removal)

Filtering adalah tahap mengambil kata-kata penting dari hasil token dengan menggunakan algoritma stoplist (membuang kata kurang penting) atau wordlist (menyimpan kata penting).

```python
stop_id = stopwords.words("indonesian")
```

```python
tokens = [token for token in tokens if token not in stop_id]
tokens
```

## d. Spell check (bahasa indonesia)
sering kali user dalam mengetik menggunakan kata singkatan dan ada juga yang typo atau salah ketik. spell check ini berguna untuk memperbaiki penulisan agar tidak ada kata-kata yang berarti sama dibaca oleh mesin dengan artian yang berbeda sumber: [Medium @yasuradb](https://medium.com/@yasirabd/spell-check-indonesia-menggunakan-pre-trained-fasttext-model-14e90a3f1ac0)

## e. Stemming 
Stemming adalah proses menghilangkan infleksi kata ke bentuk dasarnya, namun bentuk dasar tersebut tidak berarti sama dengan akar kata (root word). Misalnya kata “mendengarkan”, “dengarkan”, “didengarkan” akan ditransformasi menjadi kata “dengar”.

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
stemming = []
for i in tokens:
    x = stemmer.stem(i)
    stemming.append(x)
```

# 3. rancangan feature yang dipilih agar klasifikasi berjalan dengan baik 

Dari data text chat tersebut saya akan merancang feature dengan memilih Text chat kalimat dengan panjang kata minimal 2, jika kita hanya menggunakan satu kata saja akan sulit menemukan tema dari sebuah text chat tersebut. maka dari itu saya akan menggunakan vectorizing TFIDF N-gram, 
Dengan TF-IDF kita dapat memahami
   - seberapa berguna sebuah kata bagi sebuah kalimat (yang membantu kita memahami pentingnya sebuah kata dalam sebuah kalimat).
   - seberapa berguna sebuah kata bagi sebuah dokumen (yang membantu kita memahami kata-kata penting dengan lebih banyak frekuensi dalam sebuah dokumen).
   - membantu kami mengabaikan kata-kata yang salah eja (menggunakan teknik n-gram)
   <br>sumber: https://medium.com/analytics-vidhya/tf-idf-term-frequency-technique-easiest-explanation-for-text-classification-in-nlp-with-code-8ca3912e58c3
   
karena data yang akan menggunakan data berupa data text chat besar kemungkinan kita menemukan akan banyak kata yang salah eja atau salah tulis, maka tenik TFIDF n-gram lah yang paling cocok untuk digunakan.
