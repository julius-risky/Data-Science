```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

from sklearn.decomposition import PCA, NMF

from luwiji.dimensionality_reduction import illustration
```

# Import Data

```python
df = pd.read_csv("data/lfw_people.csv")
df.head()
```

# Dataset Spliting

```python
X = df.drop(columns="name").values
y = df.name.values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Visualize data

```python
idx = 2

plt.imshow(X_train[idx].reshape(87,65), cmap='gray')
plt.title(y_train[idx])
plt.axis('off');
```

# Decide n_components

```python
pca = PCA().fit(X_train)
```

```python
plt.plot(pca.explained_variance_ratio_.cumsum())
plt.xlabel("n_components")
plt.ylabel("Culmulative Explained Variance")
```

# Feature Extraction melalui dim Reduction

```python
pca = PCA(250, whiten=True)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
print(X_train.shape) #hiden layer
print(X_train_pca.shape) #wheigt / code
print(pca.components_.shape) #fitur
```

# Visualization fitur (component)

```python
plt.figure(figsize=(15,8))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(pca.components_[i].reshape(87,65), cmap = "gray")
    plt.title(f"PCA komponent ke- {i+1}")
    plt.axis("off");
```

# Recunstruct Image

```python
X_train_inv = pca.inverse_transform(X_train_pca)
```

```python
idx = 2
plt.figure(figsize=(8,5))

plt.subplot(121)
plt.imshow(X_train[idx].reshape(87,65), cmap="gray")
plt.title(f"asli {y_train[idx]}")
plt.axis('off');

plt.subplot(122)
plt.imshow(X_train_inv[idx].reshape(87,65), cmap="gray")
plt.title(f"inv {y_train[idx]}")
plt.axis('off');

```
