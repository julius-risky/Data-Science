```python
import nltk
nltk.download("stopwords")
nltk.download("punkt")
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
text ="Halo, nama saya juki. Saya makan bakso harga Rp 15.000 hari ini. Kamu mau?"
```

## Normalization

```python
text = text.lower()
text
```

## Tokenization

memecah belah kalimat dengan token yang kita mau

```python
sent_tokenize(text)
```

```python
tokens = word_tokenize(text)
```

## Punctuation Removal / Alphanumeric Cleansing

```python
punctuation
```

```python
tokens = [token for token in tokens if token not in punctuation]
tokens
```

```python
tokens = [token for token in tokens if token.isalnum()]
tokens
```

## Stop words removal

```python
stop_id = stopwords.words("indonesian")
```

```python
tokens = [token for token in tokens if token not in stop_id]
tokens
```
